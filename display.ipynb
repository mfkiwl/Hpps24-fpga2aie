```python
import nibabel as nib
import os
import numpy as np
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def display_image_with_resolution(image_path):
    """
    Opens an image file, displays it, and prints its resolution.

    Args:
        image_path (str): The path to the image file.
    """
    try:
        # Check if the image file exists
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"Error: Image file not found at {image_path}")

        # Open the image using Pillow
        img = Image.open(image_path)

        # Get image resolution (width, height)
        width, height = img.size

        print(f"Displaying image: {os.path.basename(image_path)}")
        print(f"Resolution: {width} x {height} pixels")

        # Display the image using Matplotlib
        plt.figure(figsize=(8, 6)) # Set a reasonable figure size
        plt.imshow(img)
        plt.title(f"Image: {os.path.basename(image_path)}\nResolution: {width} x {height} pixels")
        plt.axis('off') # Hide axes for cleaner image display
        plt.show()

    except FileNotFoundError as e:
        print(e)
    except Image.UnidentifiedImageError:
        print(f"Error: Could not identify image file at {image_path}. It might be corrupted or not a supported image format.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")



image_to_display = "./PNG/axial_slice_0009.png"  

display_image_with_resolution(image_to_display)
```
